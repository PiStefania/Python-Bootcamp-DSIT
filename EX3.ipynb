```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("housing_california.csv", sep=',')
```

```python
# 1
data.describe()
```

```python
# 2
null_columns = data.columns[data.isnull().any()]
null_rows = data[null_columns].isnull().sum(numeric_only=True).sum()
print("Total number of null rows: " + str(null_rows))
```

```python
# 3
avg = data.mean()
filled_data = data.fillna(avg)
filled_data
```

```python
# 4
filled_data.describe()
```

```python
# 5
unique_values = data.ocean_proximity.unique()
ocean_proximity_mapping = { unique_values[i] : i+1 for i in range(0, len(unique_values) ) }
#print(ocean_proximity_mapping)
mapping_data = data.applymap(lambda s: ocean_proximity_mapping.get(s) if s in ocean_proximity_mapping else s)
mapping_data
```

```python
# 6
mapping_data.describe()
```

```python
# 7.1
plt.scatter(mapping_data.median_house_value, mapping_data.ocean_proximity)
plt.xlabel("Median House Value")
plt.ylabel("Ocean Proximity")
plt.title("Median House Value vs Ocean Proximity")
```

```python
# 7.2
plt.scatter(mapping_data.median_income, mapping_data.ocean_proximity)
plt.xlabel("Median Income")
plt.ylabel("Ocean Proximity")
plt.title("Median Income vs Ocean Proximity")
```

```python
# 8
rows_median_house_stmt = len(data[data["median_house_value"] > 500000])
print("Rows with median_house_value > 500000 are: " + str(rows_median_house_stmt))
```

```python
# 9
rows_housing_median_stmt = len(data[data["housing_median_age"] > 40])
print("Rows with housing_median_age > 40 are: " + str(rows_housing_median_stmt))
```
